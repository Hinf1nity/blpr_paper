```python
from ultralytics import YOLO
import cv2
import numpy as np

model = YOLO('char_weights.pt')
Image_Path = "tests/Test_Model_7.png"
```

```python
def Plates_Number(Model_Number):
    if 0 <= Model_Number <= 9:
        return chr(ord('0') + Model_Number)
    elif 10 <= Model_Number <= 35:
        return chr(ord('A') + (Model_Number - 10))
    else:
        return ''
```

```python
image = cv2.imread(Image_Path)
Resized_Image = cv2.resize(image, (700, 500), interpolation=cv2.INTER_AREA)
Results = model(Resized_Image)

Characters = []
```

```python
for Detections in Results:
    for Box in Detections.boxes:
        x1, y1, x2, y2 = map(int, Box.xyxy[0])
        Confidence = Box.conf[0].item()
        cls = int(Box.cls[0].item())
        Detected_Character = Plates_Number(cls)
        
        Characters.append((x1, Detected_Character))
        
        cv2.rectangle(Resized_Image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        Text = f"{Detected_Character}: {Confidence:.2f}"
        cv2.putText(Resized_Image, Text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
```

```python
Characters.sort(key=lambda x: x[0])
Plate_Number = "".join([char for _, char in Characters])
print(f"PLATE NUMBER: {Plate_Number}")
```
